9) Se cuenta con un RDD con info sobre patentamientos de autos con la siguiente informacion: (patente, marca, modelo, version, tipo_vehiculo, provincia, fecha), donde tipo_vehiculo indica si la unidad patentada es auto, pickup, camion o moto. **Se pide generar un programa en PySpark que indique la marca y modelo del auto mas patentado por tipo de vehiculo en Buenos Aires en Abril de 2017.**

```python
import random
from fechasRandom import *
from datetime import *

fechaInicial = '2017-01-01'
fechaFinal = '2017-05-01'
tipoVehiculos = ['Auto','Pickup','Camion','Moto']
provincias = ['Buenos Aires','Cordoba','Santa Fe','Neuquen','Rio Negro',
              'Santa Cruz']
patentamientos = []

for i in range(100):
    patente = i
    marca = "Marca {}".format(random.randint(1,3))
    modelo = "Modelo {}".format(random.randint(1,3))
    version = "Version {}".format(random.randint(1,3))
    tipoVehiculo = random.choice(tipoVehiculos)
    provincia = random.choice(provincias)
    fecha = random_date(fechaInicial, fechaFinal)
    patentamientos.append((patente,marca,modelo,version,tipoVehiculo,provincia,fecha))
    
patentamientosRDD = sc.parallelize(patentamientos)
```

```python
patentamientosRDD.collect()
```

**Se pide generar un programa en PySpark que indique la marca y modelo del auto mas patentado por tipo de vehiculo en Buenos Aires en Abril de 2017.**

```python
#me quedo con los patentamientos de Buenos aires en abril de 2017
patentamientosBuenosAires = patentamientosRDD.filter(lambda x: x[5] == 'Buenos Aires')
patentamientosAbril = patentamientosBuenosAires.filter(lambda x: datetime.strptime(x[6],'%Y-%m-%d').month == 4)
patentamientosAbril2017 = patentamientosAbril.filter(lambda x: datetime.strptime(x[6],'%Y-%m-%d').year == 2017)

patentamientosAbril2017 = patentamientosAbril2017.map(lambda x: ((x[4],x[1],x[2]),1))
patentamientosAbril2017 = patentamientosAbril2017.reduceByKey(lambda x,y: x+y)

patentamientosAbril2017 = patentamientosAbril2017.map(lambda x:(x[0][0],(x[0][1],x[0][2],x[1])))
patentamientosAbril2017 = patentamientosAbril2017.reduceByKey(lambda x,y: x if x[2] > y[2] else y)
patentamientosAbril2017.collect()
```
