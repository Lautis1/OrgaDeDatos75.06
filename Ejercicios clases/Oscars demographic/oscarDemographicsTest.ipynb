```python
import pandas as pd
import numpy as np

OscarsDataframe = pd.read_csv('Oscars-demographics-DFE.csv', encoding='latin-1')
OscarsDataframe
```

**Para ir probando, me quiero quedar con la gente que nacio en el año 1890. Necesito convertir el tipo de la columna *date_of_birth* en *datetime*. Además voy a eliminar las columnas que no necesito, quedandome solo con *birthplace, date_of_birth, award, movie, person*.**

```python
MyDataframe = OscarsDataframe[['birthplace','date_of_birth','award','movie','person']]
MyDataframe
```

```python
MyDataframe.loc[:,'date_of_birth'] = pd.to_datetime(MyDataframe['date_of_birth'], errors='coerce')
MyDataframe
```

**Hasta ahora tengo el DataFrame filtrado por las columnas que quiero. Ahora, filtro por año de nacimiento 1890**


```python
MyDataframe1890 = MyDataframe[MyDataframe['date_of_birth'].dt.year == 1890]
MyDataframe1890
```

**PROBANDO EL groupBy()**

```python
MyDataframeTest = MyDataframe.groupby('movie')
MyDataframeTest.get_group('Bad Girl')
```
