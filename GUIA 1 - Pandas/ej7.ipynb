Dada la exitosa convocatoria de los juegos olimpicos de la juventud por parte del publico, sus organizadores realizan distintos analisis para planificar las jornadas finales del certamen. Es por ello que cuentan con informacion en los siguientes archivos.csv:
-eventos.csv (id_evento, fecha, id_locacion, nombre_evento, id_categoria_deportiva, cantidad_espectadores),
-locacion.csv (id_locacion, nombre, capacidad, capacidad_extendida, sede, latitud, longitud),
-categorias_deportivas.csv (id_categoria_deportiva, nombre, año_de_adopcion)

El primer archivo cuenta con info de eventos, indicando la fecha YYY-mm-dd hh:mm:ss, lugar donde ocurrio (id_locacion) y la cantidad de espectadores q asistieron. Ademas se aprota info sobre la categoria deportiva a la cual pertenece el evento.
Por otro lado se tiene info sobre las distintas locaciones en las sedes del certamen en las que ocurrieron los eventos. Contamos con info de su capacidad total de espectadores asi como de su capacidad extendida (cuantos asientos extra se pueden brindar).
**Se desea obtener:
    A. Nombre de la sede que acumulo la mayor cantidad de espectadores en eventos durante el certamen del 14 al 15 de ocutbre inclusive. Esto es de vital importancia para distribuir el merchandising oficial del evento, para las fechas finales.
    B. Nombre del evento y nombre de la categoria deportiva de aquellos eventos cuya cantidad de espectadores supero la capacidad de la locacion, mas alla de la capacidad extendida. Esto es de vital importancia para detectar problemas de seguridad o si es necesario realizar algun cambio de locacion.**
    

```python
import pandas as pd
import random
import numpy as np

def random_dates(start, end, n=10):
    start_u = start.value//10**9
    end_u = end.value//10**9
    return pd.to_datetime(np.random.randint(start_u, end_u, n), unit='s')
```

```python


dataFrameEventos = pd.DataFrame(columns = ['id_evento','fecha','id_locacion','nombre_evento','id_categoria_deportiva','cantidad_espectadores'])
dataFrameLocacion = pd.DataFrame(columns = ['id_locacion','nombre','capacidad','capacidad_extendida','sede','latitud','longitud'])
dataFrameCategoriasDeportivas = pd.DataFrame(columns = ['id_categoria_deportiva','nombre','año_de_adopcion'])

fechaInicial = pd.to_datetime('2019-09-14')
fechaFinal = pd.to_datetime('2019-10-15')

for i in range(150):
    fecha = random_dates(fechaInicial, fechaFinal,1)[0]
    dataFrameEventos = dataFrameEventos.append({
        'id_evento' : i,
        'fecha' : "{}-{}-{} {}:{}:{}".format(fecha.year, fecha.month, fecha.day,
                                             fecha.hour, fecha.minute, fecha.second
                                        ),
        'id_locacion' : i,
        'nombre_evento' : "Evento {}".format(i),
        'id_categoria_deportiva' : i,
        'cantidad_espectadores' : random.randint(0,600)
    }, ignore_index = True)

    
    dataFrameLocacion = dataFrameLocacion.append({
        'id_locacion' : i,
        'nombre' : 'Nombre sede {}'.format(i),
        'capacidad' : random.randint(0,400),
        'capacidad_extendida' : random.randint(0,200),
        'sede' : i,
        'latitud' : random.randint(-90,90),
        'longitud' : random.randint(-180,180)
    }, ignore_index = True)
    
    dataFrameCategoriasDeportivas = dataFrameCategoriasDeportivas.append({
        'id_categoria_deportiva' : i,
        'nombre' : "Categoria {}".format(i),
        'año_de_adopcion' : random.randint(1990,2005)
    }, ignore_index = True)
    
dataFrameEventos['fecha'] = pd.to_datetime(dataFrameEventos['fecha'], format = "%Y-%m-%d %H:%M:%S")


```

**A. Nombre de la sede que acumulo la mayor cantidad de espectadores en eventos durante el certamen del 14 al 15 de ocutbre inclusive. Esto es de vital importancia para distribuir el merchandising oficial del evento, para las fechas finales.**

```python
eventosOctubre = dataFrameEventos[dataFrameEventos['fecha'].dt.month == 10]
eventosOctubre = eventosOctubre[eventosOctubre['fecha'].dt.day.isin([14,15])]
eventosOctubrePorSede = dataFrameLocacion.merge(eventosOctubre, on = 'id_locacion')
maxEspectadores = eventosOctubrePorSede['cantidad_espectadores'].max()
eventosOctubrePorSede[eventosOctubrePorSede['cantidad_espectadores'] == maxEspectadores]['nombre'].to_string()[1:]

```

 **B. Nombre del evento y nombre de la categoria deportiva de aquellos eventos cuya cantidad de espectadores supero la capacidad de la locacion, mas alla de la capacidad extendida. Esto es de vital importancia para detectar problemas de seguridad o si es necesario realizar algun cambio de locacion.**

```python
eventosPorLocacion = dataFrameEventos.merge(dataFrameLocacion, on = 'id_locacion')
eventosQueSuperanCapacidad = eventosPorLocacion[eventosPorLocacion.cantidad_espectadores > (eventosPorLocacion.capacidad + eventosPorLocacion.capacidad_extendida)]
eventosMerged = dataFrameCategoriasDeportivas.merge(eventosQueSuperanCapacidad, on = 'id_categoria_deportiva')
eventosMerged.rename(columns = {'nombre_x' : 'nombre_categoria'}, inplace = True)
#me quedo solo con las columnas que quiero, usando filter(..., axis = 1)
eventosMerged.filter(['nombre_evento','nombre_categoria','cantidad_espectadores','capacidad','capacidad_extendida'])
```
