El GCPD recolecta la info de casos policiales que acontecen en ciudad gotica. Esta info se encuentra guardad en un DF con el formato (fecha, id_caso, descripcion, estado_caso, categoria, latitud, longitud). Los posibles estados son 1: caso abierto, 2: caso resuelto, 3:cerrado sin resolver. Las fechas se encuentra en formayo YYYY-MM-DD. Por otro lado, el comisionado Gordon guarda un registro detallado sobre en cuales casos fue activada la batise単al para pedir ayuda del vigilante , Batman. Esta info se encuentra en otro DF con formato (id_caso, respuesta), siendo campo respuesta 1: si la se単al tuvo respuesta positiva, 0: respuesta negativa de parte de batman. El sector encargado de las estadisticas oficiales del GCPD quiere con esta info, analizar los sig casos:
A) Tasa de resolucion de casos de la fuerza policial por categoria de caso (casos en los q no participio batman)
B) Tasa de resolucion de casos con la ayuda de Batman (casos en los que fue llamado con batise単al, participio en la resolucion)
C) indicar el mes del a単o pasado en el que batman tuvo mayor participacion en la investigacion de casos.

```python
import pandas as pd
import random

dataFrameGCPD = pd.DataFrame(columns = ['fecha','id_caso','descripcion','estado_caso','categoria','latitud','longitud'])
dataFrameGordon = pd.DataFrame(columns = ['id_caso','respuesta'])

for id_caso in range(30):
    dataFrameGCPD = dataFrameGCPD.append({
        'fecha' : "{}-{}-{}".format(random.randint(2017,2020), random.randint(1,12), random.randint(1,31)),
        'id_caso' : id_caso,
        'descripcion' : "Descripcion caso {}".format(id_caso),
        'estado_caso' : ['caso abierto','caso resuelto','caso cerrado sin resolver'][random.randint(0,2)],
        'categoria' : "Categoria caso {}".format(id_caso),
        'latitud': random.randint(-90,90),
        'longitud' : random.randint(-180,180)
    }, ignore_index = True)
    
    dataFrameGordon = dataFrameGordon.append({
        'id_caso' : id_caso,
        'respuesta' : random.randint(0,1)
    }, ignore_index = True)
```

```python
# Tasa de resolucion de casos de la GCPD por categoria, en los q no participo batman
#supongo que es la cantidad de casos resueltos
dataFrameSinBatman = dataFrameGordon[dataFrameGordon['respuesta'] == 0]
dataFrameCasosCerrados = dataFrameGCPD[dataFrameGCPD['estado_caso'] != 'caso abierto']
dataFrameSinBatman = dataFrameCasosCerrados.merge(dataFrameSinBatman, on = 'id_caso')
#agrupo por categoria y calculo el % de casos resueltos

dataFrameSinBatman.groupby('categoria')['estado_caso'].agg(lambda x: (x != 'caso abierto').mean()).rename('tasa de resolucion')
```

```python
#Tasa de resolucion de casos con ayuda de batman
dataFrameConBatman = dataFrameGordon[dataFrameGordon['respuesta'] == 1]
dataFrame = dataFrameGCPD.merge(dataFrameConBatman, on = 'id_caso')
dataFrame.estado_caso.apply(lambda x: (x != 'caso abierto')).mean()
```

```python
#Indicar mes del 2019 en el q batman tuvo mayor participacion investigando casos
dataFrame['fecha'] = pd.to_datetime(dataFrame['fecha'], format = '%Y-%m-%d')
#obtengo los casos de 2019 en los q participo batman
dataFrame2019 = dataFrame[dataFrame['fecha'].dt.year == 2019]

dataFrame2019['fecha'].dt.month.value_counts().idxmax()
```
