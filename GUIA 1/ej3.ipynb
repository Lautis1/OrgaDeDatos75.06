CONSIGNA:
      Dado un DF con formato (Codigo producto, Codigo fabricante, mes, ventas mensuales):
-A) Generar un df conteniendo (codigo producto, promedio de ventas)
-B) Genear un df conteniendo codigo de producto y minimo,maximo  promedio de ventas por producto mensual(en filas)
-C) mismo que B pero por columnas (codigo, minimo, maximo, promedio) mensual
-D) minimo, maximo y promedio de ventas por producto, con formato:
(tipo de calculo, codigo producto1, codigo producto2,......) donde tipo de calculo es minimo, maximo y promedio de ventas
-E) qu√© productos son provistos por distintos fabricantes, indicando true/false con el formato:
(codigo de productos (en filas los distintos codigos), codigo fabricante 1,........)

```python
import pandas as pd
import numpy as np
import random

#creo un dataframe de prueba
dataFrame = pd.DataFrame(columns = ['Codigo producto','Codigo fabricante','Mes','Ventas mensuales'])
for i in range(3):
    for j in range(1,13):
        dataFrame = dataFrame.append({
            'Codigo producto': int(i),
            'Codigo fabricante' : int(i) + random.randint(1,15),
            'Mes' : int(j),
            'Ventas mensuales' : int(random.randint(0,500))
        }, ignore_index = True)
dataFrame['Codigo producto'] = dataFrame['Codigo producto'].astype(int)
dataFrame['Codigo fabricante'] = dataFrame['Codigo fabricante'].astype(int)
dataFrame['Mes'] = dataFrame['Mes'].astype(int)
dataFrame['Ventas mensuales'] = dataFrame['Ventas mensuales'].astype(int)
dataFrame
```

```python
#A)
promedioDeVentasPorCodigo =dataFrame.groupby('Codigo producto').agg({'Ventas mensuales' : 'mean'})
promedioDeVentasPorCodigo.columns = ['Promedio de ventas']
promedioDeVentasPorCodigo
```

```python
#B)
maxMinPromedioPorProducto = dataFrame.groupby('Codigo producto').agg({'Ventas mensuales': ['min','max','mean']})
maxMinPromedioPorProducto.columns = ['Minimo ventas mensual','Maximo ventas mensual','Promedio ventas mensual']
maxMinPromedioPorProducto.stack().to_frame()
```

```python
#C)
maxMinPromedioPorProducto
```

```python
#D) transpongo el anterior
maxMinPromedioPorProducto.T
```

```python
#E)
x = dataFrame[['Codigo producto', 'Codigo fabricante']]
x = x.assign(valor = True) #asigna una nueva columna 'valor' con True
#uso pivot_table en vez de pivot porque hay mas de 1 entrada index
x = x.pivot_table(index = 'Codigo producto',columns = 'Codigo fabricante', values = 'valor')
# como me quedan NaN dando vueltas, los reemplazo por False con fillna()
x.fillna(False)
```
