9) UN importante broker de compra y venta de vehiculos online se encuentra dando sus primeros pasos en la preparacion de su algoritmo de pricing, es por eso que se encuentra generando algunos featuers iniciales para experimentar con distintos algoritmos de machine learning. Para ello cuenta con un archivo con informacion de todas las transacciones que tuvo en su primer año de operacion en el formato (transaction_id, timestamp, vehicle_model_id, price).
Por otro lado cuenta con informacion que fue extrayendo a partir de scrapping durante el ultimo año en el formato (timestamp, source, vehicle_model_id, price). La info puede venir de multiples fuentes (source), que pueden ser por ejemplo distintos sitios de marketplace o clasificados.
Luego de un intenso trabajo previo ha podido unificar los modelos de vehiculos que utiliza para sus transacciones con la info q ha podido obtener de otros competidores mediante scrapping. Muchos de los modelos disponibles en la informacion de scrapping no han sido aun comercializados por la empresa, pero se sabe que se cuenta con precios scrappeados de todos los modelos que se vendieron.

**Se pide generar un dataframe que tenga el siguiente formato en pandas:
(vehicle_mode_id, ext_mean_price, ext_std_price, int_mean_price,int_std_price) donde:
-mean_price: promedio de precio para ese vehiculo
-std_price: desvio estandar del precio para ese vehiculo.
Los prefijos ext e int indican q deben ser calculado sobre respectivamente datos externos (scrapping) y datos internos(transacciones)**

```python
import pandas as pd
import numpy as np
import random

def random_dates(start, end, n=10):
    start_u = start.value//10**9
    end_u = end.value//10**9
    return pd.to_datetime(np.random.randint(start_u, end_u, n), unit='s')

fechaInicial = pd.to_datetime('2019-01-01')
fechaFinal = pd.to_datetime('2019-12-31')

dataFrameInfoTransacciones = pd.DataFrame(columns = ['transaction_id','timestamp','vehicle_model_id','price'])
dataFrameScrapping = pd.DataFrame(columns = ['timestamp','source','vehicle_model_id','price'])

for id_transaccion in range(30):
    dataFrameInfoTransacciones = dataFrameInfoTransacciones.append({
        'transaction_id' : id_transaccion,
        'timestamp' : random_dates(fechaInicial, fechaFinal, 1)[0],
        'vehicle_model_id' : random.randint(1,20),
        'price' : random.uniform(150000,750000)
    }, ignore_index = True)
    
    dataFrameScrapping = dataFrameScrapping.append({
        'timestamp' : random_dates(fechaInicial, fechaFinal, 1)[0],
        'source' : 'Source {}'.format(id_transaccion),
        'vehicle_model_id' : random.randint(1,20),
        'price' : random.uniform(150000,750000)
    }, ignore_index = True)
```

**Se pide generar un dataframe que tenga el siguiente formato en pandas: (vehicle_mode_id, ext_mean_price, ext_std_price, int_mean_price,int_std_price) donde: -mean_price: promedio de precio para ese vehiculo -std_price: desvio estandar del precio para ese vehiculo. Los prefijos ext e int indican q deben ser calculado sobre respectivamente datos externos (scrapping) y datos internos(transacciones)**

```python
dfMergeado = dataFrameInfoTransacciones.merge(dataFrameScrapping, on = 'vehicle_model_id', how = 'outer')
#outer hace la UNION de los DF, no la interseccion como inner

#elimino columnas q no necesito: transaction_id, timestamp_x, timestamp_y,source
dfMergeado = dfMergeado.drop(columns = ['transaction_id','timestamp_x','timestamp_y','source'])

#renombro columnas
dfFinal = dfMergeado.rename(columns = {'price_x':'price_int', 'price_y':'price_ext'})

#agrupo por numeor de vehiculo
dfFinal = dfFinal.groupby('vehicle_model_id').agg({'price_int' : ['mean','std'], 'price_ext' : ['mean','std']})
dfFinal.columns = ['ext_mean_price','ext_std_price','int_mean_price','int_std_price']
dfFinal
```
